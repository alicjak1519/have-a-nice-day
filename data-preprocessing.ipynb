```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
sample_sub = pd.read_csv("data/example_submission.csv")
test = pd.read_csv("data/icml_face_data.csv")
```

```python
for i in range(10):
    img = test.iloc[i][2] # first image
    val = img.split(" ")
    x_pixels = np.array(val, 'float32')
    x_pixels /= 255
    x_reshaped = x_pixels.reshape(48,48)
    plt.imshow(x_reshaped, cmap= "gray", interpolation="nearest")
    plt.title(test.iloc[i][0])
    plt.axis("off")
    plt.show()
```
